```python
import pandas as pd
import numpy as np
df = pd.read_csv ("/Users/danort/Desktop/Ironhack labs/lab-customer-analysis-round-2/files_for_lab/csv_files/marketing_customer_analysis.csv")
df.head()

df.columns = df.columns.str.upper().str.replace(' ', '_').str.replace('[^a-zA-Z0-9_]', '')
df
                
```

```python
#Which columns are numerical

numerical_columns = df.select_dtypes("number")
numerical_columns
```

```python
#which ones are categorical 
categorical_columns = df.select_dtypes("object")
categorical_columns
```

```python
#Checking and deal with NaN values.

# This can give us an idea of the types and the amount of Nans per column:

df.info()
```

```python
# This way we can sum up the Nans per columns:
df.isna().sum()
```

```python
df['STATE'].value_counts()
```

```python
# filling Nans with a categorical value:

df['STATE'] = df['STATE'].fillna('F')
df['RESPONSE'] = df['RESPONSE'].fillna('F')
df["MONTHS_SINCE_LAST_CLAIM"] = df["MONTHS_SINCE_LAST_CLAIM"].fillna('F')
df["NUMBER_OF_OPEN_COMPLAINTS"] = df["NUMBER_OF_OPEN_COMPLAINTS"].fillna('F')
df['VEHICLE_CLASS'] = df['VEHICLE_CLASS'].fillna('F')
df['VEHICLE_SIZE'] = df['VEHICLE_SIZE'].fillna('F')
df['VEHICLE_TYPE'] = df['VEHICLE_TYPE'].fillna('F')

df.isna().sum()
```

```python
df['EFFECTIVE_TO_DATE'] = pd.to_datetime(df['EFFECTIVE_TO_DATE'])
df
```

```python
df['EFFECTIVE_TO_DATE']
```

```python
#Extracting the months from the dataset and store in a separate colum

df['MONTH'] = df['EFFECTIVE_TO_DATE'].dt.month
df['MONTH']
```

```python
#filter the data to show only the information for the first quarter 
first_quarter_data = df[(df['EFFECTIVE_TO_DATE'].dt.month >= 1) & (df['EFFECTIVE_TO_DATE'].dt.month <= 3)]
first_quarter_data
```
